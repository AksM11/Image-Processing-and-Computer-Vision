```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Initialize a blank display[2D array]
out = np.zeros((300, 500), dtype=np.uint8)
# The center coordinates of all 15 dots for one digit are stored in a list.
dots = []
x=25
for j in range(3):
  y=50;
  for i in range(5):
    dots = dots + [[x,y]]
    y=y+50
  x=x+50
```

```python
#Dots:
#0 5 10
#1 6 11
#2 7 12
#3 8 13
#4 9 14

#Initialize the black dots for each digit
dig_arr = [0]*10
dig_arr[0] = [6,7,8]
dig_arr[1] = [0,1,2,3,4,10,11,12,13,14]
dig_arr[2] = [1,6,8,13]
dig_arr[3] = [1,3,6,8]
dig_arr[4] = [3,4,5,6,8,9]
dig_arr[5] = [3,6,8,11]
dig_arr[6] = [6,8,11]
dig_arr[7] = [1,2,3,4,6,7,8,9]
dig_arr[8] = [6,8]
dig_arr[9] = [3,6,8]
```

```python
inp = input("Enter your roll number: ")
# Generate image digit by digit

for j in range(2):
  left = 500-217*(j+1) #distance of digit from left border
  digit = int(inp[-j-1])

  for i in range(len(dots)):
      dots[i][0] = dots[i][0] + left
      if i not in dig_arr[digit]:
        cv2.circle(out, tuple(dots[i]), 25, 255, -1);
      dots[i][0] = dots[i][0] - left

# Show as grayscale image.
plt.imshow(out, 'gray')
```
